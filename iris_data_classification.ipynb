#Split tha Data into training and testing parts


```python
X= df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

#Preprocessing the Data


*   Scaling down the values in the feature data which it will make the training of the model efficient.


```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

#Choosing the model


*   The model selection is the most crucial imp part in the machine learning
*   We use Logistic Regression Model


```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

#Evaluate the model


```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
confusion = confusion_matrix(y_test, y_pred)
confusion
```

```python
classification_rep = classification_report(y_test, y_pred)
print(classification_rep)
```

```python
result = pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
```

```python
check=[]
for i in range(len(result)):
  if result.iloc[i,0]==result.iloc[i,1]:
    check.append("Correct")
  else:
    check.append("Incorrect")
```

```python
result['Check']=check
```

```python
result.head()
```

```python
px.scatter(result,x='Actual',y='Predicted',color='Check',template='plotly_dark')
```
